```python
selected_columns = data[['Y2017', 'Area']]
area_sum_2017 = selected_columns.groupby('Area')['Y2017'].sum()
area_with_highest_sum = area_sum_2017.idxmax()

# The 'area_with_highest_sum' variable now contains the area with the highest sum in 'Y2017'
print("Area with the highest sum in Y2017:", area_with_highest_sum)
```
